```python
import pandas as pd
```

```python
vendedores_df = pd.read_excel('pandas 7 - datos de vendedores.xlsx')
```

```python
vendedores_df.head(10)
```

```python
region_asignada = []

for item in vendedores_df['EMAIL']:
    if '.ar' in item:
        region_asignada.append('Argentina')
    elif '.br' in item:        
        region_asignada.append('Brasil')
    elif '.uy' in item:
        region_asignada.append('Uruguay')
    elif '.py' in item:
        region_asignada.append('Paraguay')
    elif '.ec' in item:
        region_asignada.append('Ecuador')
    elif '.cl' in item:
        region_asignada.append('Chile')
    elif '.co' in item:
        region_asignada.append('Colombia')
    else:
        region_asignada.append('Región no Asignada')
```

```python
vendedores_df['REGION'] = region_asignada
vendedores_df.tail(10)
```

```python
vendedores_df.info()
```

```python
vendedores_df['TELEFONO'] = vendedores_df['TELEFONO'].astype(str)
vendedores_df.info()
```

```python
i = 0

nuevos_telefonos = []

for item in vendedores_df['EMAIL']:    
    if '.ar' in item:        
        nuevo_telefono = '+54-' + vendedores_df.iloc[i]['TELEFONO']        
        nuevos_telefonos.append(nuevo_telefono)
    elif '.br' in item:        
        nuevo_telefono = '+55-' + vendedores_df.iloc[i]['TELEFONO']
        nuevos_telefonos.append(nuevo_telefono)
    elif '.uy' in item:
        nuevo_telefono = '+598-' + vendedores_df.iloc[i]['TELEFONO']
        nuevos_telefonos.append(nuevo_telefono)
    elif '.py' in item:
        nuevo_telefono = '+595-' + vendedores_df.iloc[i]['TELEFONO']
        nuevos_telefonos.append(nuevo_telefono)
    elif '.ec' in item:
        nuevo_telefono = '+593-' + vendedores_df.iloc[i]['TELEFONO']
        nuevos_telefonos.append(nuevo_telefono)
    elif '.cl' in item:
        nuevo_telefono = '+56-' + vendedores_df.iloc[i]['TELEFONO']        
        nuevos_telefonos.append(nuevo_telefono)
    elif '.co' in item:
        nuevo_telefono = '+57-' + vendedores_df.iloc[i]['TELEFONO']
        nuevos_telefonos.append(nuevo_telefono)
    i +=1
```

```python
vendedores_df['TELEFONO'] = nuevos_telefonos
```
