# <center>El Método groupby()</center>

```python
import pandas as pd
```

```python
vendedores_df = pd.read_excel('pandas 12 - datos de vendedores.xlsx')
```

```python
vendedores_df.head(10)
```

```python
vendedores_df.tail(10)
```

```python
vendedores_df.info()
```

```python
vendedores_df.groupby(["REGION ASIGNADA"]).count()
```

```python
vendedores_df.groupby(["REGION ASIGNADA"])["VENTAS"].count()
```

```python
vendedores_df.groupby(["REGION ASIGNADA"])["VENTAS"].mean()
```

```python
vendedores_df.groupby(["REGION ASIGNADA"])["VENTAS"].sum()
```

```python
vendedores_df.groupby(["REGION ASIGNADA","CALIFICACION DADA POR LOS CLIENTES"])["VENTAS"].sum()
```

```python
vendedores_df.groupby(["REGION ASIGNADA","CALIFICACION DADA POR LOS CLIENTES"])["VENTAS"].mean()
```

```python
vendedores_df.groupby(vendedores_df.FECHA_INGRESO.dt.year).count()
```
