```python
import pandas as pd
```

# Ciudades de Brasil

```python
ciudades1_df = pd.DataFrame({
                    'Ciudad':['Curitiba','Porto Alegre','Santos','Fortaleza','Florianópolis'],
                    'Estado':['Paraná', 'Rio Grande do Sul','Sao Paulo','Ceará','Santa Catarina'],
                    'Fundación':[1963,1772,1546,1726,1673],
                    'Población':['1,864,416','1,409,939','419,757','2,669,340', '508,826'],
                    'Superficie (km2)':[432,496,281,313,675]
                            })
ciudades1_df.head()
```

```python
ciudades2_df = pd.DataFrame({
                    'Ciudad':['Florianópolis','Torres','Rio de Janeiro','Recife','Maceió','Curitiba','Porto Alegre'],
                    'Estado':['Santa Catarina', 'Rio Grande do Sul','Rio de Janeiro','Pernambuco','Alagoas','Paraná','Rio Grande do Sul'],
                    'Fundación':[1673,1878,1565,1527,1815,1963,1772],
                    'Temperatura':[28,30,32,38,37,32,33],
                    'Prefijo Telefónico':[48,51,21,81,82,41,51]
                            })
ciudades2_df.head(7)
```

```python
ciudades_full_df = pd.concat([ciudades1_df, ciudades2_df], ignore_index=True)
ciudades_full_df.head(12)
```

# Inner

```python
ciudades_full_inner_df = pd.merge(ciudades1_df, ciudades2_df)
ciudades_full_inner_df.head()
```

# Left

```python
ciudades_full_left_df = pd.merge(ciudades1_df, ciudades2_df, how='left', on='Ciudad')
ciudades_full_left_df.head(10)
```

# Right

```python
ciudades_full_right_df = pd.merge(ciudades1_df, ciudades2_df, how='right', on='Ciudad')
ciudades_full_right_df.head(10)
```

# Outer

```python
ciudades_full_outer_df = pd.merge(ciudades1_df, ciudades2_df, how='outer', on='Ciudad')
ciudades_full_outer_df.head(10)
```
