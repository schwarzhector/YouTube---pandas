# Eliminar Valores Faltantes

```python
vendedores_df.head(8)
```

```python
v_sin_nan_df = vendedores_df.dropna()
v_sin_nan_df.head()
```

```python
vendedores_df.dropna(how = 'all', inplace=True)
print(vendedores_df.shape)
vendedores_df.head(8)
```

```python
v_sin_nan_df = vendedores_df.dropna(subset=['REGION'])
print(v_sin_nan_df.shape)
v_sin_nan_df.head(7)
```
