```python

equipos_df['PJ'].plot(kind='bar',
                      grid=True,
                      xlabel='Selecciones',
                      ylabel='Partidos Jugados',
                      title='Estadísticas de la Copa Mundial de Fútbol'
                     )
```

```python
equipos_df['PJ'].plot(kind='bar',
                      grid=True,
                      xlabel='Selecciones',
                      ylabel='Partidos Jugados',
                      title='Estadísticas de la Copa Mundial de Fútbol',
                      fontsize = 15
                     )
```

```python
equipos_df['PJ'].plot(kind='bar',
                      grid=True,
                      xlabel='Selecciones',
                      ylabel='Partidos Jugados',
                      title='Estadísticas de la Copa Mundial de Fútbol',
                      fontsize = 15,
                      figsize=(12,6)
                     )
```

```python
equipos_df['PJ'].plot(kind='bar',
                      grid=True,
                      xlabel='Selecciones',
                      ylabel='Partidos Jugados',
                      title='Estadísticas de la Copa Mundial de Fútbol',
                      fontsize = 15,
                      figsize=(12,6),
                      color='darkgreen'
                     )
```

```python
equipos_df['PJ'].plot(kind='barh',
                      grid=True,
                      xlabel='Selecciones',
                      ylabel='Partidos Jugados',
                      title='Estadísticas de la Copa Mundial de Fútbol',
                      fontsize = 15,
                      figsize=(6,12),
                      color='darkorange'
                     )
```

```python
equipos_df.plot.scatter(x='PG',
                        y='GF',
                        grid=True,
                        figsize=(10,10), 
                        xlabel='Partidos Ganados',
                        ylabel='Goles a Favor',
                        color='navy')
```

```python
equipos_df.plot.scatter(x='PJ',
                        y='Dif.',
                        grid=True,
                        figsize=(10,10), 
                        xlabel='Partidos Jugados',
                        ylabel='Diferencia de Goles',
                        color='indigo'
                       )
```
